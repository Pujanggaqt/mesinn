```python
# Import library yang diperlukan
import os
import urllib.request
import pandas as pd
import os
os.makedirs("datasets", exist_ok=True)
```

```python
# Tentukan URL dataset baru
NEW_DATASET_URL = "https://raw.githubusercontent.com/Pujanggaqt/mesinn/main/onlinefoods.csv"
```

```python
# Fungsi untuk mengunduh dataset baru
def fetch_new_data(dataset_url=NEW_DATASET_URL):
    csv_path = os.path.join("datasets", "onlinefoods.csv")
    urllib.request.urlretrieve(dataset_url, csv_path)
```

```python
# Terapkan fungsi untuk mengunduh dataset baru
fetch_new_data()
```

```python
# Fungsi untuk memuat dataset baru dengan nama variabel "food"
def load_new_data(csv_path):
    return pd.read_csv(csv_path)
```

```python
# Memuat dataset baru dengan nama variabel "food"
food = load_new_data("datasets/onlinefoods.csv")
```

```python
# Menampilkan informasi dataset baru
print("Informasi Dataset:")
print(food.info())
```

```python
# Menampilkan jumlah nilai "
print("\nJumlah Nilai pada Variabel 'onlinefoods.csv':")
print(food["Age"].value_counts())
```

```python
# Menampilkan statistik deskriptif masing-masing data
food.describe()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
food.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Menampilkan data dengan missing value
sample_incomplete_rows = food[food.isnull().any(axis=1)]
sample_mv = sample_incomplete_rows
sample_mv
```

```python

#menghapus data yang memiliki missing value
sample_dropna = sample_mv
sample_dropna.dropna(subset=["Gender"])
```

```python
# Menampilkan data setelah penghapusan missing value
sample_dropna
```

```python
# Menggunakan OrdinalEncoder untuk mengubah kategori menjadi nilai ordinal
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder_food = OrdinalEncoder()

sample_cat_encoded_food = ordinal_encoder_food.fit_transform(food)
sample_cat_encoded_food
```

```python
# Menampilkan kategori ordinal
ordinal_encoder_food.categories_
```

```python
# Melihat contoh data
sample_norm_food = food[["Family size"]]
sample_norm_food
```

```python
# Menggunakan MinMaxScaler untuk normalisasi data
from sklearn.preprocessing import MinMaxScaler
min_max_scaler_food = MinMaxScaler(feature_range=(0, 100))
minmax_sample_food = min_max_scaler_food.fit_transform(sample_norm_food)
minmax_sample_food
```
